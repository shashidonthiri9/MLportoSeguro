```python
training_data_majority=training_data[training_data.target==0]
training_data_minority=training_data[training_data.target==1]
```

```python
from sklearn.utils import resample
training_data_majority_downsample=resample(training_data_majority,replace=False,n_samples=100000,random_state=123)
```

```python
training_data_downsample=pd.concat([training_data_majority_downsample,training_data_minority])
```

```python
training_data_downsample.target.value_counts()
training_data_downsample.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(training_data_downsample.drop('target',axis=1))
```

```python
scaled_features = scaler.transform(training_data_downsample.drop('target',axis=1))
```

```python
training_feat = pd.DataFrame(scaled_features,columns=test_data.columns)
training_feat.head()
```

```python
#training_data_downsample.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_validation, y_train, y_validation = train_test_split(scaled_features, training_data_downsample['target'], test_size=0.20, random_state=101)
```

```python
#NOW TRAIN THE MODEL
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_validation)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_validation,predictions))
```

```python
print(confusion_matrix(y_validation,predictions))
```

```python
from sklearn.metrics import accuracy_score, roc_curve, auc
```

```python
fpr,tpr,thresholds = roc_curve(y_validation, predictions)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr,label='ROC curve (area = %0.2f)' % roc_auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
